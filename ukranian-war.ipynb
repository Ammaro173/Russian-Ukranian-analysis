### Loading the Head(5) of the data frame of Russia_losses_equipment file

```python
df_equ =pd.read_csv('../input/2022-ukraine-russian-war/russia_losses_equipment.csv')
df_equ.head(5)
```

```python
df_equ.info()
```

# Seaborn

```python
import seaborn as sns
# sns.set() # alias for set_theme()
sns.set_theme()
```

### Day After Day The Russian_Ukranian was is showing more and more losses of equipments such as TANKS , Air craft , Helicopters and APC(Guns) as the next Plots will show the losses of each type

```python
sns.relplot(x="day", y="tank",data=df_equ)
```

```python

sns.relplot(x="day", y="aircraft",data=df_equ)
```

### Based on data below , you can see the losses of equipment are increasing linearly , daily!!; From whether its from aircraft equips, to APC(Guns) equips. 

> so they are  really consuming alot of resources daily!!

```python
col = [ 'aircraft', 'helicopter', 'tank', 'APC'
      ]
sns.set_theme()
for i in col:
    sns.lmplot(x="day", y=i,data=df_equ, palette="muted")
```

### Loading the Head(5) of the data frame of Russia_losses_personnel file

```python
df_pers =pd.read_csv('../input/2022-ukraine-russian-war/russia_losses_personnel.csv')
df_pers.head(5),df_pers.info()

```


**Now plotting the relation of the daily personnel loses from the personnel data Frame !**
